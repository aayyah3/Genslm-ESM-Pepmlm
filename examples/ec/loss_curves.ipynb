```python
from __future__ import annotations
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl


def plot_loss_curves(
    csv_file_paths: list[str],
    labels: list[str],
    colors: list[str],
    field: str = "eval_loss",
    y_label: str = "Evalulation loss",
    title: str = "",
    save_path: str = "",
) -> None:

    # Set Avenir font and larger font size
    mpl.rcParams["font.family"] = "Avenir"
    mpl.rcParams["font.size"] = 18

    plt.figure(figsize=(10, 6))

    for csv_file_path, label, color in zip(csv_file_paths, labels, colors):
        # Load data from CSV file using Pandas
        df = pd.read_csv(csv_file_path)

        # Extract columns
        steps, plot_data = df["step"], df[field]

        plt.plot(
            steps, plot_data, label=label, linestyle="-", linewidth=4, alpha=1, c=color
        )

    plt.title(title)
    plt.xlabel("Step")
    plt.ylabel(y_label)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    # Save or display the plot
    if save_path:
        plt.savefig(save_path, dpi=600, bbox_inches="tight")
    plt.show()
```

# Evaluation Loss

```python
plot_loss_curves(
    csv_file_paths=[
        "loss_curves/ec_aminoacid_8m.csv",
        "loss_curves/ec_aminoacid_35m.csv",
        "loss_curves/ec_aminoacid_150m.csv",
        "loss_curves/ec_aminoacid_650m.csv",
    ],
    labels=["8M", "35M", "150M", "650M"],
    colors=["#a6cee3", "#1f78b4", "#b2df8a", "#33a02c"],
    title="EC Amino Acid Loss Curves",
    save_path="figures/ec_aminoacid_loss.png",
)
```
