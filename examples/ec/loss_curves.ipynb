```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_loss_curves(csv_file_path):
    # Load data from CSV file using Pandas
    df = pd.read_csv(csv_file_path)

    # Extract columns
    epochs = df['epoch']
    training_losses = df['loss']
    evaluation_losses = df['eval_loss']

    plt.figure(figsize=(10, 6))
    
    # Plot training loss
    plt.plot(epochs, training_losses, label='Training Loss', marker='o', linestyle='-', color='blue')
    
    # Plot evaluation loss
    plt.plot(epochs, evaluation_losses, label='Evaluation Loss', marker='o', linestyle='-', color='orange')
    
    plt.title('Training and Evaluation Loss Curves')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    
    # Save or display the plot
    plt.savefig('loss_curves.png')
    plt.show()
```

```python
plot_loss_curves("loss_curves/ec_aminoacid_8m.csv")
```

```python

from __future__ import annotations
import pandas as pd
import matplotlib.pyplot as plt

def plot_loss_curves(
    csv_file_paths: list[str], 
    labels: list[str], 
    colors: list[str],
    field: str = "eval_loss",
    y_label: str = "Evalulation loss",
    title: str = "",
    save_path: str = "",
) -> None:
    
    plt.figure(figsize=(10, 6))

    for csv_file_path, label, color in zip(csv_file_paths, labels, colors):
        # Load data from CSV file using Pandas
        df = pd.read_csv(csv_file_path)

        # Extract columns
        steps, plot_data = df['step'], df[field]

        plt.plot(steps, plot_data, label=label, marker='o', linestyle='-', alpha=0.7, c=color)

    plt.title(title)
    plt.xlabel('Step')
    plt.ylabel(y_label)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    # Save or display the plot
    if save_path:
        plt.savefig(save_path, dpi=600, bbox_inches='tight')
    plt.show()
```

```python
plot_loss_curves(
    csv_file_paths=[
        "loss_curves/ec_aminoacid_8m.csv",
        "loss_curves/ec_aminoacid_35m.csv",
        "loss_curves/ec_aminoacid_150m.csv",
        "loss_curves/ec_aminoacid_650m.csv",
    ], 
    labels=["8M", "35M", "150M", "650M"],
    colors=["#ffffcc", "#a1dab4", "#41b6c4", "#225ea8"],
    title="EC Amino Acid Loss Curves"
)
```
